```python
'''Todos os dias, o nosso sistema atualiza as vendas do dia anterior.\n",
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos
vendidos no dia anterior",
"E-mail da diretoria: seugmail+diretoria@gmail.com<br>\n",
"Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing\n",
"Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado"'''

import pyautogui
import pyperclip
import time
import pandas as pd

#AQUI TEM TODOS OS CODIGOS DO PYAUTOGUI: https://pyautogui.readthedocs.io/en/latest/quickstart.html
```

```python
#tempo/delay
pyautogui.PAUSE = 1

# Passo 1: Entrar no sistema (no nosso caso, entrar no link)
pyautogui.hotkey('ctrl', 't') #abrindo uma aba
pyperclip.copy('https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing') #copiando link
pyautogui.hotkey("ctrl", 'v') #colando link
pyautogui.press('enter') #entrando no sistema
time.sleep(5) #O sistema espera carregar por 5s

# Passo 2: Navegar até o local do relatório (entrar na pasta Exportar)
pyautogui.click(x=436, y=295, clicks=2)

# Passo 3: Fazer o download do relatório
pyautogui.click(x=411, y=404) #clicar no arquivo
pyautogui.click(x=1719, y=195)#clicar nos 3 pontos
pyautogui.click(x=1464, y=616)##clicar em fazer download
```

```python
# Passo 4: Importar a base de vendas pro Python
tabela = pd.read_excel(r'C:\Users\PICHAU\Pictures\Vendas - Dez.xlsx') #armazenar o valor/variavel #diretorio/caminho/local
display(tabela)

# Passo 5: Calcular o faturamento e quantidade de produtos vendidos
faturamento = tabela['Valor Final'].sum()
qtd_produtos = tabela['Quantidade'].sum()
```

```python
# Passo 6: Enviar por e-mail o resultado
#Abrir email
pyautogui.hotkey('ctrl', 't')
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl", 'v')
pyautogui.press('enter')
time.sleep(5)
#clicar no botao escrever
pyautogui.click(x=1051, y=210)
#preencher destino
pyautogui.write('brochjosias@gmail.com')
pyautogui.press('tab')#seleciona o email
pyautogui.press('tab')#muda para o campo de assunto
#preencher o assunto
pyautogui.write('Relatório de Vendas')
pyautogui.press('tab')#mudar para o campo do corpo do email

#escrever o email
texto = f"""
Prezados, bom dia!

O faturamento de ontem foi de R$: {faturamento}
A quantidade de produtos vendidos foi de: {qtd_produtos}

Abs
Josias Broch
"""
#f = formatar com variaveis
pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')
#clicar em enviar
pyautogui.hotkey("ctrl", "enter")
```
